```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
anime = pd.read_csv('../input/anime-recommendations-database/anime.csv')
rating = pd.read_csv('../input/anime-recommendations-database/rating.csv')
```

```python
data = pd.merge(rating, anime, on = 'anime_id')
data = data.drop('rating_y', axis = 1)
data = data.drop('episodes', axis = 1)
data.rename(columns = {'rating_x':'rating'}, inplace = True)
data = data[data.rating != -1]
data
```

```python
Data = pd.DataFrame(columns = ['Anime', 'Genre', 'Streaming type', 'Rating', 'Views'])
Data['Anime'] = data.name
Data['Genre'] = data.genre
Data['Streaming type'] = data.type
Data['Rating'] = data.rating
Data['Views'] = data.members

Data = Data.drop_duplicates(subset='Anime')
Data.dropna(inplace = True)
Data = Data.sort_values(by = 'Views', ascending=False)
```

```python
AVAnimeRate = data.drop(columns = ['genre', 'user_id', 'anime_id', 'type'])

AVAnimeRate.rename(columns = {'name':'Anime'}, inplace = True)
AVAnimeRate.rename(columns = {'rating':'Rating'}, inplace = True)
AVAnimeRate.rename(columns = {'members':'Views'}, inplace = True)

AVAnimeRate = AVAnimeRate.groupby('Anime', as_index=False).mean().round(2)
AVAnimeRate = AVAnimeRate.sort_values(by = 'Rating', ascending = False)
AVAnimeRate = AVAnimeRate[AVAnimeRate['Views'] > 10000]
```

```python
TopMostPopular = Data
TopMostPopular = TopMostPopular.drop(columns=['Streaming type', 'Genre', 'Rating'])
```

# Top Anime Series

```python
TopAnimeToWatch = pd.merge(AVAnimeRate, TopMostPopular, on = 'Views')
TopAnimeToWatch.sort_values(by = 'Views', ascending = False)
TopAnimeToWatch = TopAnimeToWatch.drop(columns = ['Anime_y'])
TopAnimeToWatch.rename(columns = {'Anime_x':'Anime name'}, inplace = True)
TopAnimeToWatch
```

# Most Watched Anime Series

```python
TopViews = TopAnimeToWatch.nlargest(n=30, columns = ['Views'])
TopViews = TopViews.reset_index(drop=True)
TopViews
```

```python
TopRated = TopAnimeToWatch.nlargest(n=30, columns = ['Rating'])
TopRated
```
